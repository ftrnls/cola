```python
import networkx as nx
import matplotlib.pyplot as plt

# Buat graph
G = nx.DiGraph()

# tambahkan node

G.add_node("Mulai")
G.add_node("Shopee FOOD")
G.add_node("Pilih menu")
G.add_node("Masukan alamat")
G.add_node("Pilih metode pembayaran")
G.add_node("pesan sekarang")
G.add_node("Tunggu kurir makanan datang")
G.add_node("selesai")

#tambahkan edge(hubungan antar code)
G.add_edge("Mulai", "Shopee FOOD")
G.add_edge("Shopee FOOD", "Pilih menu")
G.add_edge("Pilih menu", "Masukan alamat")
G.add_edge("Masukan alamat", "Pilih metode pembayaran")
G.add_edge("Pilih metode pembayaran","pesan sekarang")
G.add_edge("pesan sekarang", "Tunggu kurir makanan datang")
G.add_edge("Tunggu kurir makanan datang","selesai")

# Gambar graph
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=700, node_color="skyblue", font_size=8, font_color="black", font_weight="bold", arrowsize=26)

# tampilkan diagram
plt.show()
```
