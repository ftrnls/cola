```python
import networkx as nx
import matplotlib.pyplot as plt

# Buat graph
G = nx.DiGraph()

# tambahkan node
G.add_node("Mulai")
G.add_node("pilih lokasi penjemputan")
G.add_node("pilih tujuan")
G.add_node("pilih Ride Type")
G.add_node("konfirmasi perjalanan")
G.add_node("pembayaran")
G.add_node("pengemudi Ditugaskan")
G.add_node("perjalanan sedang Berlangsung")
G.add_node("selesai")

#tambahkan edge(hubungan antar code)
G.add_edge("Mulai", "pilih lokasi penjemputan")
G.add_edge("pilih lokasi penjemputan", "pilih tujuan")
G.add_edge("pilih tujuan","pilih Ride Type")
G.add_edge("pilih Ride Type","konfirmasi perjalanan")
G.add_edge("konfirmasi perjalanan","pembayaran")
G.add_edge("pembayaran","pengemudi Ditugaskan")
G.add_edge("pengemudi Ditugaskan","perjalanan sedang Berlangsung")
G.add_edge("perjalanan sedang Berlangsung","perjalanan Lengkap")
G.add_edge("perjalanan Lengkap","selesai")

# Gambar graph
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=700, node_color="skyblue", font_size=8, font_color="black", font_weight="bold", arrowsize=26)

# Tampilkan diagram
plt.show()
```
