```python
import networkx as nx
import matplotlib.pyplot as plt

# Buat graph
G = nx.DiGraph()

# tambahkan node

G.add_node("BUKA gojek")
G.add_node("klik sendok dan garpu")
G.add_node("pilih kaybun")
G.add_node("pilih menu")
G.add_node("dimsum goreng")
G.add_node("pilih Alamat")
G.add_node("Pilih metode pembayaran")
G.add_node("pesanan masuk")
G.add_node("menunggu driver")
G.add_node("selesai")

#tambahkan edge(hubungan antar code)
G.add_edge("BUKA gojek", "klik sendok dan garpu")
G.add_edge("klik sendok dan garpu","pilih kaybun")
G.add_edge("pilih kaybun", "pilih menu")
G.add_edge("pilih menu", "dimsum goreng")
G.add_edge("dimsum goreng", "pilih Alamat")
G.add_edge("pilih Alamat", "Pilih metode pembayaran")
G.add_edge("Pilih metode pembayaran", "pesanan masuk")
G.add_edge("pesanan masuk", "menunggu driver")
G.add_edge("menunggu driver","selesai")

# Gambar graph
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=700, node_color="skyblue", font_size=8, font_color="black", font_weight="bold", arrowsize=26)

# tampilkan diagram
plt.show()
```
